```python
data = pd.read_csv("../data/emt-data.csv")
data.shape
```

```python
data.head(3)
```

### Clean

```python
data = data[~data["LINEAS"].isnull()]
data["PARTICULA"] = data["PARTICULA"].fillna("")
```

```python
data["DIRECCION"] = data["TIPOVIA"].str.strip() + " " + data["PARTICULA"].str.strip() + " " + data["NOMBREVIA"].str.strip() + " " + data["NUMEROPORTAL"].str.strip()
data = data[~data["DIRECCION"].isnull()]
```

```python
columns = [
    "CODIGOESTACION", "CODIGOEMPRESA", "DENOMINACION", "LINEAS", 
    "DIRECCION", "CODIGOPOSTAL", "DISTRITO",
]
data_s = data[columns]
```

### Transform

```python
data_s["LINEAS"] = data_s["LINEAS"].str.split(",")
```

```python
data_s_tabular = data_s.explode(column="LINEAS")
```

### Personalize

```python
for col in data_s_tabular.columns:
    data_s_tabular[col] = data_s_tabular[col].astype(str)
    data_s_tabular[col] = data_s_tabular[col].str.strip()
data_s_tabular.dtypes
data_s_tabular["DIRECCION"] = data_s_tabular["DIRECCION"].replace(r"\s+", " ")
data_s_tabular = data_s_tabular.rename(columns={
    "CODIGOESTACION": "codigo_estacion",
    "CODIGOEMPRESA": "codigo_parada",
    "DENOMINACION": "nombre_parada",
    "LINEAS": "linea",
    "DIRECCION": "direccion_parada",
    "CODIGOPOSTAL": "cp_parada",
    "DISTRITO": "distrito_parada"
})
```

### Transform lineas

```python
consecutive_pairs = lambda ele: [(ele[i], ele[i + 1]) for i in range(len(ele) - 1)]
```

```python
lineas_cols = ["NUMEROLINEAUSUARIO", "SENTIDO", "NUMEROORDEN", "CODIGOESTACION"]
```

```python
data_linea_s = data_linea[lineas_cols].sort_values(lineas_cols[:-1])
```

```python
data_linea_s = data_linea_s.groupby(["NUMEROLINEAUSUARIO", "SENTIDO"], as_index=False)["CODIGOESTACION"].apply(list)
```

```python
data_linea_s["CODIGOESTACION"] = data_linea_s["CODIGOESTACION"].apply(consecutive_pairs)
```

```python
data_linea_s = data_linea_s.explode("CODIGOESTACION")
```

```python
data_linea_s["from_codigo_estacion"] = data_linea_s["CODIGOESTACION"].str[0]
data_linea_s["to_codigo_estacion"] = data_linea_s["CODIGOESTACION"].str[1]
data_linea_s = data_linea_s.rename(columns={
    "NUMEROLINEAUSUARIO": "edge"
})
data_linea_s = data_linea_s[["from_codigo_estacion", "to_codigo_estacion", "edge"]]
```

```python
data_linea_s = data_linea_s.drop_duplicates(keep="first")
```

### Save
